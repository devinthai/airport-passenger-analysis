```python
import os
import pandas as pd
```

```python
dfs = []
for file in os.listdir('./data'):
    if '.csv' not in file:
        continue
    print(f"processing {file}")
    dfs.append(pd.read_csv(f'./data/{file}'))
```

```python
df = pd.concat(dfs)
```

```python
df['YEAR'].unique()
```

```python
df.to_csv('./parsed_data/combined_monthly_passenger_data.csv', index=False)
```
